## The Optical Properties of Gold
The free electrons in metals can be excited through plasma oscillations, which can be described through the Drude Lorentz Sommerfeld model in \eqref{test}. While $\omega_p$ is the plasma frequency and $\Gamma$ the damping for the oscillations.

\begin{align}
\epsilon(\omega) = 1 - \frac{\omega_p^2}{\omega^2 - i \Gamma \omega} \tag{1}
\label{test}
\end{align} 

The plasma frequency $\omega_p$ can be described with $N$, the number of electrons per volume, with eq.\ref{Eq2}.

\begin{align}
\omega_p = \frac{N e^2}{\epsilon_0 m_e} 
\label{Eq2}
\end{align} 

The dampening can be seen in eq.$\ref{manual3}$, while $v_F$ is the fermi velocity and $l_b$ the mean free path.

\begin{align}
\Gamma = \frac{v_F}{l_b}
\label{manual3}
\end{align}

Using the equations above, it can be seen that $\epsilon$ is complex, while the real parts are decribing a polarisation and the imaginary part describes an absorption. Using the maxwell relation we can define a refraction index $\tilde{n}$ (eq.\ref{manual4}).

\begin{align}
\tilde{n} = n(\omega) + i\kappa(\omega)
\label{manual4}
\end{align}

The parts of $\tilde{n}$ can also be described and are shown in eq.\ref{manual5} and eq.\ref{manual6}.

\begin{align}
n(\omega) &= \sqrt{\frac{\epsilon_{Re}(\omega)}{2} + \frac{\epsilon_{Re}(\omega)^2\epsilon_{Im}(\omega)^2}{2}}
\label{manual5} \\
i\kappa(\omega) &= \sqrt{-\frac{\epsilon_{Re}(\omega)}{2} + \frac{\epsilon_{Re}(\omega)^2\epsilon_{Im}(\omega)^2}{2}}
\label{manual6}
\end{align}

In the following the optical properties of koloidal gold particles are described. For that we lookin at the extinktion profile $\sigma_{ext}$ and is defined in eq.\ref{manual7}. $\sigma_{Abs}$ is the absorption profile and $\sigma_{Str}$ the scattering profile.

\begin{align}
\sigma_{ext}= \sigma_{Abs} +\sigma_{Str}
\label{manual7}
\end{align}

With the Maxwell equation it is possible to calculate an electromagnetic wave that is hitting the partciles. The solution are partial waves that are created by the multipole oscillations inside of the particles. For particles of the radius $R$ and $\lambda$ the wavelength of the incoming rays for the case $R << \lambda$ the extinktion profile $\sigma_{ext}$ can be calculated (eq.\ref{manual8}).

\begin{align}
\sigma_{ext}(\lambda) = \frac{24\pi^2 R^3 \epsilon^{\frac{3}{2}} \epsilon_{Im}(\lambda)}{\lambda(\epsilon_{Re}(\lambda)+ 2\epsilon_m)^2 + \epsilon_{Im}(\lambda)^2}
\label{manual8}
\end{align}

$\epsilon_m$ is the dielectricity constant of the medium for the particles. For this case the partial wave are dominant and the scattering profile $\sigma{Str}$ is low compared to $\sigma{Abs}$. Following that we can write the case for resonance, which gives us that the resonance frequency $\lambda_r$ is only dependent of $\epsilon_m$ and $\epsilon_m=n^2=1.77$

\begin{align}
(\epsilon_{Re}(\omega) +2 \epsilon_m)^2 + (\epsilon_{Im}(\omega))^2
\label{manual9}
\end{align}

\subsection{Interaction between koloidale Particles and their Medium}

By now we have not talked about the interaction with neighboring partciles. The theory of the Mie scattering only works for one partcile. So to describe an ensemble of particles we have to introduce a filling factor $f$.

\begin{align}
f= \frac{V_k}{V_l} = \frac{\pi d^3 c_0}{6} = \frac{A_k}{A_{ges}}= \frac{\pi d^2 n}{4 A_{ges}}
\label{manual10}
\end{align}

```python
Î» = np.flipud(np.array([1240/1.2, 1240/1.4, 1240/1.6, 1240/1.8, 
                        1240/2, 1240/2.1, 1240/2.2, 1240/2.4, 
                        1240/2.5, 1240/2.6, 1240/2.7, 1240/2.8, 
                        1240/2.9, 1240/3, 1240/3.1]))
n = np.flipud(np.array([0.1, 0.08, 0.08, 0.09, 0.13, 0.18, 0.24, 
                        0.5, 0.82, 1.24, 1.43, 1.46, 1.5, 1.54, 1.54]))
Îº = np.flipud(np.array([6.54, 5.44, 4.56, 3.82, 3.16, 2.84, 2.54, 
                        1.86, 1.59, 1.54, 1.72, 1.77, 1.79, 1.8, 1.81]))

def ninterp(Î»in):
    """Creates a 1D array of n values interpolated from the data
    set n.
    
    Î»in is the x-range that you want for n."""
    return np.interp(Î»in, Î», n)

def Îºinterp(Î»in):
    """Creates a 1D array of n values interpolated from the data
    set Îº.
    
    Î»in is the x-range that you want for n."""
    return np.interp(Î»in, Î», Îº)

def Ïµ(n: 'Re(n(Ï)) + Im(n(Ï))'):
    """Takes in n and Îº from Eq. 5 in the lab manual and returns
    Ïµ, which is the square of Eq. 4 from the lab manual.
    """
    return np.multiply(n, n)

# Creating a courser set of Î» to feed into the above functions. 
Î»n = np.linspace(np.min(Î»), np.max(Î»), num=300, endpoint=True)
    
# Creating a vector of complex values from the linear interpolated values.
N = np.array([np.complex(ninterp(i), Îºinterp(i)) for i in Î»n])

# Ploting the interp fucntion
fig, ax = plt.subplots(figsize=(13, 7))
ax.plot(Î»n, ninterp(Î»n), 'x', label='Interp of n')
ax.plot(Î», n, '-o', label='n')
ax.plot(Î»n, Îºinterp(Î»n), 'x', label='Interp of $\kappa$')
ax.plot(Î», Îº, '-o', label='n')
ax.grid(True)
ax.legend()
ax.set_title('Comparing the linear interp to the data') 
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Permittivity (F/m)')
ax.axhline(color='k')
plt.show()
    
# Ploting the Ïµ fucntion
fig, ax = plt.subplots(figsize=(13, 7))
ax.plot(Î»n, Ïµ(N).real, label='Re($\epsilon$)')
ax.plot(Î»n, Ïµ(N).imag, label='Im($\epsilon$)')
ax.grid(True)
ax.legend()
ax.set_title('Plots of $\epsilon$') 
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Permittivity (F/m)')
ax.axhline(color='k')
plt.show()

# Parametric plot of the fucntion
fig, ax = plt.subplots(figsize=(13, 7))
ax.plot(Ïµ(N).real, Ïµ(N).imag, 'x')
ax.grid(True)
ax.set_title('Phase plot of $\epsilon$') 
ax.set_xlabel('Re($\epsilon$)')
ax.set_ylabel('Im($\epsilon$)')
ax.axhline(color='k')
ax.axvline(color='k')

# Add labels to points
ax.text(Ïµ(N[0]).real - 2, Ïµ(N[0]).imag + 0.1, np.round(Î»n[0]))
ax.text(Ïµ(N[-1]).real - 2, Ïµ(N[-1]).imag + 0.1, np.round(Î»n[-1]))
    
    
plt.show()
```

```python
Ïµm = 1.77 # For Water

def Î(Ïµ: 'Permittivity'):
    """Returns Î: Particle polarizability 
    Eq. 17 from the lab manual"""
    return (Ïµ - Ïµm) / (Ïµ + 2 * Ïµm)

def Ïµeff(Î: 'Particle polarizability', f):
    """Returns Ïµeff: Effective Permittivity.
    Eq. 17 from the lab manual"""
    return Ïµm * (1 + 2 * f * Î) / (1 - f * Î)

# Ploting the fucntion
fig, ax = plt.subplots(figsize=(13, 7))
ax.plot(Î»n, Ïµeff(Î(Ïµ(N)), 0.2).real, label='Re($\epsilon_{eff}$)')
ax.plot(Î»n, Ïµeff(Î(Ïµ(N)), 0.2).imag, label='Im($\epsilon_{eff}$)')
ax.plot(Î»n, np.abs(Ïµeff(Î(Ïµ(N)), 0.2)), label='$|\epsilon_{eff}|$')
ax.grid(True)
ax.legend()
ax.set_title('Plots of $\epsilon_{eff}$') 
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Effective Permittivity (F/m)')
ax.axhline(color='k')
plt.show()

# Parametric plot of the fucntion
fig, ax = plt.subplots(figsize=(13, 7))
ax.plot(Ïµeff(Î(Ïµ(N)), 0.2).real, Ïµeff(Î(Ïµ(N)), 0.2).imag, '-o')
ax.grid(True)
ax.set_title('Phase plot of $\epsilon_{eff}$') 
ax.set_xlabel('Re($\epsilon_{eff}$)')
ax.set_ylabel('Im($\epsilon_{eff}$)')
ax.axhline(color='k')
ax.axvline(color='k')

# Add labels to points
ax.text(Ïµeff(Î(Ïµ(N[0])), 0.2).real, Ïµeff(Î(Ïµ(N[0])), 0.2).imag, np.round(Î»[0]))
ax.text(Ïµeff(Î(Ïµ(N[-1])), 0.2).real, Ïµeff(Î(Ïµ(N[-1])), 0.2).imag, np.round(Î»[-1]))
    
plt.show()
```

We want to find the resoance condition by looking at:
\begin{align}
| \epsilon + 2 \epsilon_m | = min
\end{align}

```python
def Ïµeffmin(Ïµm):
    return [np.min(np.abs(Ïµeff(Î(Ïµ(N)) + i, 0.2))) for i in Ïµm]

Î»min = Î»n[np.argmin(np.abs(Ïµeff(Î(Ïµ(N)) + Ïµm, 0.2)))]
Ïµeffmin0 = np.min(np.abs(Ïµeff(Î(Ïµ(N)) + Ïµm, 0.2)))

print('The Î» and Ïµ values that solve the above fucntion:')
Î»min, Ïµeffmin0

Ïµmrange = np.linspace(0, 20, num=100)

# Plot of the fucntion
fig, ax = plt.subplots(figsize=(13, 7))
ax.plot(Ïµmrange, Ïµeffmin(Ïµmrange), '-x')
ax.grid(True)
ax.set_title('Minimum of $\epsilon_{eff}$') 
ax.set_xlabel('$\epsilon_m$')
ax.set_ylabel('Minimum of $\epsilon_{eff}$ (F/m)')
ax.axhline(color='k')
ax.axvline(color='k')
```

If we range over $\epsilon_m$ we can see the the minimum of $\epsilon_{eff}$ has a maximum near 5. 

```python
def Î³(Îº, Î»):
    """Eq. 8 from manual"""
    return np.divide(4*np.pi*Îº, Î»)

def I(Î³, D=1, I0=1):
    """Eq. 7 from manual"""
    return I0 * np.exp(-D * Î³)

# Ploting the fucntion
fig, ax = plt.subplots(figsize=(13, 7))
ax.plot(Î»n, Î³(Îºinterp(Î»n), Î»n), label='$\gamma$')
ax.grid(True)
ax.legend()
ax.set_title('Plots of $\gamma$') 
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('Extinction Coefficient')
plt.show()
```

## Task 5: Soak Time Analysis

Three samples were prepared by soaking them in the gold solution for three different periods [30, 75, 220] mins. We expect that the absorption peak to be more defined for longer soak times as the surface will accumulate more Au nano particles.

Below we see that hypothesis is clearly born out by the data. For the 30min sample there is almost no signal from the Au nano particles on the surface. There is a slight wavelength shift in the max values, it seems slightly lower for lower gold saturations. However, this could be driven by the large vertical shift, almost as if there is a very broad peak running the length of the window. If we were to subtract the linear offset we might be able to account for the peak shifting. 

```python
# Find max values
Î»max75, Imax75 = np.round(data[3][data[3]['count'] == 
                                  data[3]['count'].max()].values[0], decimals=2)
Î»max220, Imax220 = np.round(data[2][data[2]['count'] == 
                                    data[2]['count'].max()].values[0], decimals=2)

# Ploting the function
fig, ax = plt.subplots(figsize=(13, 7))
ax.plot(data[1]['energy'], data[1]['count'], label='30min')
ax.plot(data[3]['energy'], data[3]['count'], 
        label=f'75min AIR Max at: ({Î»max75},  {Imax75})')
ax.plot(data[2]['energy'], data[2]['count'], 
        label=f'220min AIR Max at: ({Î»max220},  {Imax220})')
ax.grid(True)
ax.legend()
ax.set_title('Plots of $I$') 
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('I')
ax.axhline(color='k')
```

## Task 6: Different Mediums

Below we took measurements for 7 different samples, all of which were left in the Au solution for 75min then placed in various mediums for the measurement. Each fluid has a different index of refraction, which we expect to effectively change the dielectric constant and shift the peaks. 

```python
# Find max values
Î»maxGLY, ImaxGLY = np.round(data[4][data[4]['count'] == 
                                    data[4]['count'].max()].values[0], decimals=2)
Î»maxH20, ImaxH20 = np.round(data[5][data[5]['count'] == 
                                    data[5]['count'].max()].values[0], decimals=2)
Î»maxNIT, ImaxNIT = np.round(data[6][data[6]['count'] == 
                                    data[6]['count'][0:-20].max()].values[0], decimals=2)
Î»maxPENT, ImaxPENT = np.round(data[7][data[7]['count'] == 
                                      data[7]['count'].max()].values[0], decimals=2)
Î»maxPROP, ImaxPROP = np.round(data[8][data[8]['count'] == 
                                      data[8]['count'].max()].values[0], decimals=2)

# Ploting the data
fig, ax = plt.subplots(figsize=(13, 7))
ax.plot(data[3]['energy'], data[3]['count'], 
        label=f'AIR Max at: ({Î»max75},  {Imax75})')
ax.plot(data[4]['energy'], data[4]['count'], 
        label=f'GLY Max at: ({Î»maxGLY},  {ImaxGLY})')
ax.plot(data[5]['energy'], data[5]['count'], 
        label=f'H20 Max at: ({Î»maxH20},  {ImaxH20})')
ax.plot(data[6]['energy'], data[6]['count'], 
        label=f'NIT Max at: ({Î»maxNIT},  {ImaxNIT})')
ax.plot(data[7]['energy'], data[7]['count'], 
        label=f'PENT Max at: ({Î»maxPENT},  {ImaxPENT})')
ax.plot(data[8]['energy'], data[8]['count'], 
        label=f'PROP Max at: ({Î»maxPROP},  {ImaxPROP})')
ax.grid(True)
ax.legend()
ax.set_title('Plots of $I$') 
ax.set_xlabel('Wavelength (nm)')
ax.set_ylabel('I')
ax.axhline(color='k')
```

Unexpectedly, the data didn't completely reflect our expectations. For lower index of refraction fluids it was clear that $\lambda_{max}$ did shift as expected. However, there seemed to be a wall, over $n=1.34$ there was no shift to $\lambda_{max}$. This might suggest that the dielectric constant is saturating.
