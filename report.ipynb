## The Optical Properties of Gold
The free electrons in metals can be excited through plasma oscillations, which can be described through the Drude Lorentz Sommerfeld model in Eq. \ref{Eq1}. While $\omega_p$ is the plasma frequency and $\Gamma$ the damping for the oscillations.

\begin{align}
\epsilon(\omega) = 1 - \frac{\omega_p^2}{\omega^2 - i \Gamma \omega} \tag{1}
\label{Eq1}
\end{align} 

The plasma frequency $\omega_p$ can be described with $N$, the number of electrons per volume, with Eq. \ref{Eq2}.

\begin{align}
\omega_p = \frac{N e^2}{\epsilon_0 m_e}  \tag{2}
\label{Eq2}
\end{align} 

The dampening can be seen in Eq. $\ref{manual3}$, where $v_F$ is the Fermi velocity and $l_b$ is the mean free path.

\begin{align}
\Gamma = \frac{v_F}{l_b}  \tag{3}
\label{manual3}
\end{align}

From Eq. $\ref{manual3}$ it can be seen that $\epsilon$ is complex, where the real component of $\epsilon$ represents the polarisation and the imaginary component represents the absorption. Using the Maxwell relation we can define a refraction index $\tilde{n}$ (Eq. \ref{manual4}).

\begin{align}
\tilde{n} = n(\omega) + i\kappa(\omega)  \tag{4}
\label{manual4}
\end{align}

The parts of $\tilde{n}$ can also be described and are shown in Eq. \ref{manual5} and Eq. \ref{manual6}.

\begin{align}
n(\omega) &= \sqrt{\frac{\epsilon_{Re}(\omega)}{2} + \frac{\epsilon_{Re}(\omega)^2\epsilon_{Im}(\omega)^2}{2}}
\label{manual5} \tag{5} \\
i\kappa(\omega) &= \sqrt{-\frac{\epsilon_{Re}(\omega)}{2} + \frac{\epsilon_{Re}(\omega)^2\epsilon_{Im}(\omega)^2}{2}}
\label{manual6} \tag{6}
\end{align}

If we want to understand the optical properties of colloidal gold particles we need to examine at the extinction profile $\sigma_{ext}$ and is defined in Eq. \ref{manual7}. $\sigma_{Abs}$ is the absorption profile and $\sigma_{Str}$ the scattering profile.

\begin{align}
\sigma_{ext}= \sigma_{Abs} +\sigma_{Str} \tag{7}
\label{manual7}
\end{align}

With the Maxwell equation, it is possible to calculate an electromagnetic wave that is interacting with the particles. The solution can be though of as partial waves created by the multipole oscillations inside of the particles. For particles of the radius $R$ and $\lambda$ the wavelength of the incoming rays for the case $R << \lambda$ the extinction profile $\sigma_{ext}$ can be calculated (eq.\ref{manual8}).

\begin{align}
\sigma_{ext}(\lambda) = \frac{24\pi^2 R^3 \epsilon^{\frac{3}{2}} \epsilon_{Im}(\lambda)}{\lambda(\epsilon_{Re}(\lambda)+ 2\epsilon_m)^2 + \epsilon_{Im}(\lambda)^2}  \tag{8}
\label{manual8}
\end{align}

$\epsilon_m$ is the dielectric constant of the medium for the particles. For this case the partial wave is dominant and the scattering profile $\sigma{Str}$ is low compared to $\sigma{Abs}$. We can calculate the resonance condition, which gives us that the resonance frequency $\lambda_r$ is dependent on $\epsilon_m$ and $\epsilon_m=n^2=1.77$

\begin{align}
(\epsilon_{Re}(\omega) +2 \epsilon_m)^2 + (\epsilon_{Im}(\omega))^2 \tag{9}
\label{manual9} 
\end{align}

## Interaction between colloidal Particles and their Medium

By now we have not talked about the interaction with neighboring particles. The theory of the Mie scattering only works for one particle. So to describe an ensemble of particles we have to introduce a filling factor $f$.

\begin{align}
f= \frac{V_k}{V_l} = \frac{\pi d^3 c_0}{6} = \frac{A_k}{A_{ges}}= \frac{\pi d^2 n}{4 A_{ges}}  \tag{10}
\label{manual10}
\end{align}

## UV/VIS Spectroscopy

The UV/VIS spectroscopy uses electromagnetic radiation in the ultraviolet and visible light range to analyze substances in terms of their compound. It's possible to measure the absorption emissions of valence electrons during transitions excited by the incident radiation. In the measurements, the transmission or the absorption are analyzed. To get the actual concentration of the sample we use the Lambert-Beer relation, with $I_0$ is the transmission intensity of the incoming light, $I$ is the transmission intensity through the sample, $D$ the width of the sample holder.

\begin{align}
I= I_0 \exp \left( -\gamma D \right)  \tag{11}
\label{manual11}
\end{align}

The extinction constant $\gamma$ describes the decay of the intensity.

\begin{align}
\gamma(\lambda)= \frac{4 \pi \kappa(\lambda)}{\lambda} = \frac{1}{D} \ln \frac{I_0}{I}  \tag{12}
\label{manual12}
\end{align}

## AFM Spectroscopy

Atomic Force Microscopy, or AFM, uses a needle nanometers in diameter to measure the surface structure of materials. The needle is attached to a spring and the measurement is taken over the spring. The device can interact with the surface in several ways. By simply running the needle over the surface it's possible to measure the contour of the sample; however, with softer materials, this can be a destructive process. Instead, it's possible to "tap" the surface by driving the spring and measuring the dampening created by interacting with the surface of the material. We used this process in the lab to produce a measurement of the surface with embedded gold nanoparticles. 

## X-Ray Refraction


Much like AFM, it's possible to use the diffraction of X-rays off the surface of a sample to provide a mapping of the topology of the surface. The refraction index $n$ in the x-ray optic is defined with:

\begin{align}
n= \frac{1- r_0 \rho \lambda^2}{2\pi}  \tag{13}
\label{manual13}
\end{align}

We would use x-rays with $\lambda= 1.54$ for the source radiation. $r_0$ is the Thompson radius of the electron, and $\rho$ is the electron density of the sample. Using this we can measure the index of refraction of gold or glass is smaller than $1$, which is smaller than the refraction index of air. For small angles of entry $\alpha$ there is a total reflection and above the total reflection the Fresnel reflection $R_F$ is decreasing. Through interferences on the surface, we can describe the Fresnel reflection.

\begin{align}
q_z = \frac{4\pi}{\lambda} \sin{\alpha}  \tag{14}
\label{manual14}
\end{align}

The radius of the partciles is defined by:

\begin{align}
r = \frac{\pi}{\delta q_z}  \tag{15}
\label{manual15}
\end{align}

With $q_z$ being the distance between two maxima.
